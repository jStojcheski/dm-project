#### Пресметување на просечни квадратни отстапувања

```python
itypes = ['interpolation', 'mov_avg', 'season_decomp', 'season_split']

itypes_dict = {
    'interpolation': 'Interpolation', 
    'mov_avg': 'Weighted Moving Average', 
    'season_decomp': 'Seasonally Decomposed', 
    'season_split': 'Seasonally Splitted'
}
```

```python
def mse(df, x, y):
    return df[x] / 100 * ((df[x] - df[y]) ** 2)
```

```python
for itype in itypes:
    df[f'mse_{itype}'] = mse(df, 'maxRows', itype)
    
df
```

```python
errors = dict()

for itype in itypes:
    errors[itype] = dict()
    for e in set(df['TERRID']):
        errors[itype][e] = df[df['TERRID'] == e][itype].mean()
        
errors = pd.DataFrame.from_dict(errors)
errors = errors.dropna()
```

```python
for itype in itypes:
    plt.figure(figsize=(16, 4))
    ax = plt.axes()

    x = errors[itype].sort_values()
    x.index = [i[1] for i in x.index.str.split()]
    x.plot.bar(color='red', alpha=0.4, ax=ax)

#     plt.title('Просечно квадратно отстапување на секој од вработените на месечно ниво')
    plt.xlabel('Комерцијалист')
    plt.ylabel('Просечно квадратно отстапување')
    plt.ylim(40, 90)
    
    plt.legend([itypes_dict[itype]], loc=2)

    plt.savefig(f'{plot_out}/{itype}_results')
    plt.show()
```
