#### Пополнување на вредностите за month на местата каде што е извршена контрола.

Секој ден во дадена година според 4-4-5 форматот припаѓа на точно еден месец.

```python
day = list(df.Calendarday)
month = list(df.month)
```

```python
cnt = 0

for i in range(1, len(month)-1):
    if month[i] == ' ':
        if month[i-1] == month[i+1]:
            month[i] = month[i-1]
        else:
            cnt += 1

print(cnt)
```

```python
calendar = dict()

for i in range(len(month)):
    if month[i] != ' ':
        calendar[day[i]] = month[i]
```

```python
cnt = 0

for i in range(len(month)):
    if month[i] == ' ':
        if day[i] in calendar.keys():
            month[i] = calendar[day[i]]
        else:
            cnt += 1
            
print(cnt)
```

```python
df.month = pd.Series(month)
```
