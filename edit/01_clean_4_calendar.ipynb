### Замена на типот на Calendarday во pandas.Timestamp

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/Spenser_1_3_team.tsv', sep='\t')
df
```

```python
datetime = pd.to_datetime(df.Calendarday)
```

```python
df.Calendarday = datetime
df
```

Проверка дали сите вредности се точно променети.

```python
incorrect = 0

for date in df.Calendarday:
    if type(date) is not pd._libs.tslib.Timestamp:
        incorrect += 1
        
print(incorrect)
```

Export на dataset-от по оваа промена.

```python
df.to_csv('dataset/Spenser_1_4_calendar.tsv', sep='\t')
```
