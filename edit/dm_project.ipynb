#### Read Dataset

```python
df = pd.read_csv('dataset/Spenser_1.tsv', sep='\t')
df[:10]
```

```python
df.columns
```

```python
temp = df[df['APECoeficient'] != ' ']
```

```python
t1 = temp[temp['SAMEDAYVISIT'] == ' ']
```

```python
t1
```

```python
dc = dict()

for item in t1.values:
    custid = item[0]
    month = item[1].split('/')[0]
    if (custid, month) not in dc:
        dc[(custid, month)] = 1
    else:
        dc[(custid, month)] += 1
```

```python
dc
```

```python
for key in dc.keys():
    if dc[key] > 1:
        print(key)
```

```python
t1.values[0]
```

```python
dc = dict()

for item in t1.values:
    custid = item[0]
    day = int(item[1].split('/')[1])
    if (day) not in dc:
        dc[day] = 1
    else:
        dc[day] += 1
```

```python
dc
```

```python
plt.figure(figsize=(16, 9))
plt.bar(dc.keys(), dc.values())
plt.xlabel('Ден во месецот')
plt.ylabel('Број на посети')
plt.show()
```
